1. Get documents (FAQ documents in this case)

```python
import requests
```

```python
docs_url = 'https://github.com/alexeygrigorev/llm-rag-workshop/raw/main/notebooks/documents.json'
```

```python
docs_response = requests.get(docs_url)
docs_raw = docs_response.json()
```

```python
documents = []

for course in docs_raw:
    course_name = course['course']

    for document in course['documents']:
        document['course'] = course_name
        documents.append(document)
```

```python
len(documents)
documents[59]
```

2. Examine data

```python
import pandas as pd

df = pd.DataFrame(documents, columns=['course', 'section', 'question', 'text'])
```

3. Vector Spaces

- turn documents into vectors
- term-document matrix:
     - rows: documents
     - columns: words/tokens
- bag of words
    - word order is lost
    - sparse matrix
    - TF-IDF

```python
# sklearn transformer

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
cv.fit(df.text)
```

```python
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out().shape
```

```python
cv2 = CountVectorizer(min_df=5, stop_words="english")
cv2.fit(df.text)
cv2.get_feature_names_out()
```

```python
cv2.get_feature_names_out().shape
```

That looks more reasonable.

```python
X = cv2.transform(df)
```

```python
X.todense()
```

```python
df_dense = pd.DataFrame(X.todense(), columns=cv2.get_feature_names_out())
```

```python
len(df_dense)
```

```python
df_dense
```

OK, enough with the Bag of Words approach...

Time for TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(stop_words='english', min_df=5)
```

```python
X = cv.fit_transform(df.text)
```

```python
names = cv.get_feature_names_out()
```

```python
names
```

```python
df_docs = pd.DataFrame(X.toarray(), columns = names)
```

```python
df_docs.T
```

```python
df_docs.round(2)
```

```python
query = "Do I need to know python to sign up for the January course?"
```

```python
q = cv.transform([query])
q.toarray()
```
